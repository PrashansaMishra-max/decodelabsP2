# Exploratory Data Analysis (EDA)

Exploratory Data Analysis is performed to understand the distribution of variables, identify trends, detect outliers, and discover relationships between features. The insights obtained during this stage help in selecting appropriate preprocessing techniques and machine learning models.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Features:")
print(list(numerical_cols))

print("\nCategorical Features:")
print(list(categorical_cols))
```

```python
df["OrderStatus"].value_counts()
```

```python
plt.figure(figsize=(8,5))

df["OrderStatus"].value_counts().plot(kind="bar")

plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Count")

plt.show()
```
